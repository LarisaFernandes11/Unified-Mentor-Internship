### 1. Loading and Inspecting the Data
Let's start by loading the dataset into a DataFrame and inspecting it for any initial cleaning.

```python
# Importing all necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Upload amazon dataset and read the dataset

dataset=pd.read_csv("/content/amazon.csv")
dataset
```

### 2. Data Cleaning and Preprocessing

a.Handling missing values.

d.This involves checking for missing data, duplicates, and ensuring all columns have the correct data types.



```python
# Let's get some information about the dataset and understand the data

dataset.info()
```

#### Converting 'Order Date' to datetime format

```python
# Replace non-numeric values in the 'rating' column
dataset['rating'] = pd.to_numeric(dataset['rating'], errors='coerce')

# Check for any NaN values in the 'rating' column
print(dataset['rating'].isna().sum())

# Drop rows where 'rating' is NaN, or you can fill them with a default value (e.g., the mean rating)
dataset = dataset.dropna(subset=['rating'])  # Dropping NaN values

# Now convert 'rating' to float
dataset['rating'] = dataset['rating'].astype(float)

# Check the data types after conversion
print(dataset.dtypes)
```

### Step 3: Exploratory Data Analysis (EDA)
Now, let's analyze the data to derive insights:

3.1. Most Popular Products by Rating and Review Count
Analyze which products have the highest ratings and most reviews:

```python
# Top 10 products by rating
top_rated_products = dataset.sort_values(by='rating', ascending=False)[['product_name', 'rating', 'rating_count']].head(10)
print(top_rated_products)

# Top 10 products by number of ratings
most_reviewed_products = dataset.sort_values(by='rating_count', ascending=False)[['product_name', 'rating_count', 'rating']].head(10)
print(most_reviewed_products)
```

#### 3.2. Discount Analysis
Evaluate the discount percentage and how it relates to sales:

```python
# Top 10 products with the highest discount
top_discounted_products = dataset.sort_values(by='discount_percentage', ascending=False)[['product_name', 'discount_percentage', 'discounted_price', 'actual_price']].head(10)
print(top_discounted_products)
```

#### 3.3. Distribution of Ratings
Visualize the distribution of product ratings to understand customer satisfaction:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of ratings
plt.figure(figsize=(8,6))
sns.histplot(dataset['rating'], bins=10, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

### Step 4: Sentiment Analysis of Reviews
Using customer reviews, we can perform a basic sentiment analysis.

4.1. Review content for sentiment analysis:

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords and tokenizer
import nltk
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_review(text):
    # Remove special characters, numbers, and extra spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()

    # Tokenization and removing stopwords
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    return ' '.join(filtered_tokens)

# Apply preprocessing to the review content
dataset['cleaned_review'] = dataset['review_content'].apply(preprocess_review)
dataset[['review_content', 'cleaned_review']].head()
```

#### 4.2. Sentiment Scoring
Use a sentiment analysis library to score the reviews (using TextBlob or VADER):

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Apply sentiment analysis
dataset['sentiment'] = dataset['cleaned_review'].apply(get_sentiment)
dataset[['cleaned_review', 'sentiment']].head()

# Visualize the sentiment scores
plt.figure(figsize=(8,6))
sns.histplot(dataset['sentiment'], bins=20, kde=True)
plt.title('Sentiment Analysis of Customer Reviews')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

### Step 5: Product Category Analysis
We can analyze which product categories are most popular or offer the best deals:

```python
# Count of products by category
category_count = dataset['category'].value_counts().head(10)
print(category_count)

# Plot the category distribution
plt.figure(figsize=(10,6))
sns.barplot(x=category_count.values, y=category_count.index)
plt.title('Top 10 Product Categories')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```
